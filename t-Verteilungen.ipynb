### t-Verteilung vs. Normalverteilung

Die t-Verteilung beschreibt die Unsicherheit des standardisierten Mittelwerts,
wenn die Standardabweichung aus der Stichprobe geschätzt wird.
Sie hängt von den Freiheitsgraden $$df$$ ab (typisch: $$df=n-1$$).

- Bei kleinen $$df$$ hat die t-Verteilung „dickere“ Ränder (mehr Masse in den Extremen) als $$N(0,1)$$:
  Das spiegelt zusätzliche Unsicherheit bei kleinen Stichproben wider.
- Mit wachsendem $$df$$ nähert sich die t-Verteilung der Standardnormalverteilung $$N(0,1)$$ an.

Merksatz: Kleine Stichprobe ⇒ breiter/„schwerere Tails“ ⇒ größere kritische Werte.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, norm

# x-Achse
x = np.linspace(-5, 5, 2000)

# Freiheitsgrade
dfs = [1, 2, 5, 10, 30, 100]    # Eingabe der Freiheitsgrade = Anzahl der Verteilungen

plt.figure(figsize=(9, 5))

# t-Verteilungen
for df in dfs:
    plt.plot(x, t.pdf(x, df), label=f"t(df={df})")

# Normalverteilung
plt.plot(x, norm.pdf(x), label="N(0,1)", linewidth=2)

plt.title("t-Verteilung und Normalverteilung bei verschiedenen Freiheitsgraden")
plt.xlabel("x")
plt.ylabel("Dichte")
plt.legend()
plt.grid(True, alpha=0.3)

# Als PDF speichern
plt.savefig("t_verteilungen.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
